```python
# print(titers)
list_idtiter=[]
list_types=[]
list_42d=[]
list_personnr=[]
for i, row in titers.iterrows():
    print(row.samplenr)
    sample=str(row.samplenr)
    personnr=sample[0:6]
    list_personnr.append(personnr)
    if re.match(r'[0-9]{6,7}(042)', sample):
        # print(sample)
        list_idtiter.append(str(sample))
        list_42d.append(str(sample))
        sample_t= '42d'

    elif re.match(r'[0-9]{6,7}(182)', sample):
        list_idtiter.append(str(sample))
        list_6m.append(str(sample))
        sample_t= '6m'

    elif re.match(r'[0-9]{6,7}(365)', sample):
        list_idtiter.append(str(sample))
        list_12m.append(str(sample))
        sample_t= '12m'

    elif re.match(r'[0-9]{6,7}(001)', sample):
        list_idtiter.append(str(sample))
        list_prevac1.append(str(sample))
        sample_t= 'prevac1'

    elif re.match(r'[0-9]{6,7}(021)', sample):
        list_idtiter.append(str(sample))
        list_prevac2.append(str(sample))
        sample_t= 'prevac2'

    elif re.match(r'[0-9]{6,7}(049)', sample):
        list_idtiter.append(str(sample))
        list_6wpostvac.append(str(sample))
        sample_t= '6wpostvac'

    else:
        list_idtiter.append('?')
        sample_t='?'
    list_types.append(sample_t)
# print(list_personnr)

titers['idtiter'] = list_personnr

# print(list_idtiter)

titers['sample_type']= '?'

titers['sample_type']= list_types

titers

```

```python
def titersample(df, value):
    df = titers[titers['sample_type'] == str(value)].reset_index(drop=True)
    keeplist = [str(value)+ '_' + x for x in df.columns]
    # print(keeplist)
    df.columns = keeplist
    df=df.drop(columns=str(value)+'_sample_type')
    print(df)
    return df

df42d= pd.DataFrame()
df6m= pd.DataFrame()
df12m= pd.DataFrame()
dfprevac1= pd.DataFrame()
dfprevac2= pd.DataFrame()
df6wpostvac= pd.DataFrame()
dfunsure= pd.DataFrame()


df42d=titersample(df42d, '42d')
df6m= titersample(df6m, '6m')
df12m= titersample(df12m, '12m')
dfprevac1= titersample(dfprevac1, 'prevac1')
dfprevac2= titersample(dfprevac2, 'prevac2')
df6wpostvac= titersample(df6wpostvac, '6wpostvac')
dfunsure= titersample(dfunsure, '?')

# df6m = titers[titers['sample_type'] == '6m'].reset_index(drop=True)
# df12m = titers[titers['sample_type'] == '12m'].reset_index(drop=True)
# dfprevac1 = titers[titers['sample_type'] == 'prevac1'].reset_index(drop=True)
# dfprevac2 = titers[titers['sample_type'] == 'prevac2'].reset_index(drop=True)
# df6wpostvac = titers[titers['sample_type'] == '6wpostvac'].reset_index(drop=True)

df42d, df6m, df12m, dfprevac1, dfprevac2, df6wpostvac, dfunsure
```
