```python
import pandas as pd
import re
from Bio import SeqIO
import numpy as np
```

```python
def translate(seq):
	
	table = {
		'ATA':'I', 'ATC':'I', 'ATT':'I', 'ATG':'M', 'AT-': 'x', 'A--': '-',
		'ACA':'T', 'ACC':'T', 'ACG':'T', 'ACT':'T', 'AC-': 'x',
		'AAC':'N', 'AAT':'N', 'AAA':'K', 'AAG':'K', 'AA-': 'x',
		'AGC':'S', 'AGT':'S', 'AGA':'R', 'AGG':'R',	'AG-': 'x',			
		'CTA':'L', 'CTC':'L', 'CTG':'L', 'CTT':'L', 'CT-': 'x', 'C--': '-',
		'CCA':'P', 'CCC':'P', 'CCG':'P', 'CCT':'P', 'CC-': 'x',
		'CAC':'H', 'CAT':'H', 'CAA':'Q', 'CAG':'Q', 'CA-': 'x',
		'CGA':'R', 'CGC':'R', 'CGG':'R', 'CGT':'R', 'CG-': 'x',
		'GTA':'V', 'GTC':'V', 'GTG':'V', 'GTT':'V', 'GT-': 'x', 'G--': '-',
		'GCA':'A', 'GCC':'A', 'GCG':'A', 'GCT':'A', 'GC-': 'x',
		'GAC':'D', 'GAT':'D', 'GAA':'E', 'GAG':'E', 'GA-': 'x',
		'GGA':'G', 'GGC':'G', 'GGG':'G', 'GGT':'G', 'GG-': 'x',
		'TCA':'S', 'TCC':'S', 'TCG':'S', 'TCT':'S', 'TC-': 'x', 'T--': '-',
		'TTC':'F', 'TTT':'F', 'TTA':'L', 'TTG':'L', 'TT-': 'x',
		'TAC':'Y', 'TAT':'Y', 'TAA':'_', 'TAG':'_', 'TA-': 'x',
		'TGC':'C', 'TGT':'C', 'TGA':'_', 'TGG':'W', 'TG-': 'x', '---': '-'
	
	}
	protein =""


	if seq:
		for i in range(0, len(seq), 3):
			codon = seq[i:i + 3]
			if codon in table:
				protein+= table[codon]
			else:
				protein+= 'x'
		return protein
```

```python
df=pd.DataFrame()
with open('genomic_analyses//SEQ//MSA//final_seq_downsampled_229E_MSA_PROT.fasta') as file:
    for seq_rec in SeqIO.parse(file, 'fasta'):
        seq_id= seq_rec.description
        if seq_id:
            record= {'id': seq_id, 'seq': str(seq_rec.seq.lower())}
            df=df.append(record, ignore_index=True)
print(len(df), df)
```

```python
aaseqlist=[]
for i, row in df.iterrows():
    aaseq= translate(row['seq'].upper())
    aaseqlist.append(aaseq)
df['aa']=aaseqlist
df
```

```python
# listref=df['aa'][0]
# for i, row in df.iterrows():
#     for refnt in row['seq']:

locations=[]
counter=[*range(0, len(df.aa[0]))]
for i, row in df.iterrows():
    for loc, refnt, nt in zip(counter, df['aa'][0], row['aa']):
        if nt != refnt:
            locations.append(loc)





```

```python
# count how often a mutation occured
#find which items are unique
dffreq=pd.DataFrame()
used = set()
unique = [x for x in locations if x not in used and (used.add(x) or True)]
# for each unique location, count how often this location occurs in list
for item in unique:
    freq=locations.count(item)
    #percentage is part / total (total is 48)
    perc=int(freq)/(48)
    # append location and freq to list
    row= {'loc': item, 'freq': freq, 'perc':perc}
    dffreq=dffreq.append(row, ignore_index=True)
    
# print(dffreq)


listaa=df['aa'][0]
locaalist=[]
for i, row in dffreq.iterrows():
    locaa=listaa[int(row['loc'])]
    locaalist.append(locaa)
dffreq['aa']=locaalist

dffreq=dffreq[dffreq['perc']>0.5]


dffreq.to_excel('genomic_analyses\S229E_aa_mut_freq.xlsx')


```
