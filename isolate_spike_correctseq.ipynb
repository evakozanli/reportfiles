```python
import pandas as pd
import re
from Bio import SeqIO
import numpy as np
```

```python
#start with correct values


hcov= 'NL63' #choose name

## enable the correct start and end based on hcov

##for 229E spike 20570..24091
# spike_start =20569
# spike_end=24091

##for HKU1 spike 22942..27012
# spike_start= 22941
# spike_end=27012

##for OC43 spike 23643..27704
# spike_start=23642
# spike_end=27704

##for NL63 spike 20472..24542
spike_start=20471
spike_end=24542

## for SARS2 spike 21563..25384
# spike_start=21562
# spike_end=25384



df=pd.DataFrame()
with open('genomic_analyses//SEQ//full_SEQ//notambiguous//'+str(hcov)+'_notambiguouscorrect.fasta') as file:
    for seq_rec in SeqIO.parse(file, 'fasta'):
        seq_id= seq_rec.description
        if seq_id:
            record= {'id': seq_id, 'seq': str(seq_rec.seq.lower())}
            df=df.append(record, ignore_index=True)
print(len(df), df)
```

```python
df1=pd.DataFrame()
with open('genomic_analyses//'+str(hcov)+'_ref.fasta') as file:
    for seq_rec in SeqIO.parse(file, 'fasta'):
        seq_id= seq_rec.description
        if seq_id:
            record= {'id': seq_id, 'seq': str(seq_rec.seq.lower())}
            df1=df1.append(record, ignore_index=True)
print(len(df1), df1)
```

```python
df2=pd.DataFrame()
with open('genomic_analyses//SEQ//S_SEQ//notambiguous//'+str(hcov)+'_notambiguouscorrect.fasta') as file:
    for seq_rec in SeqIO.parse(file, 'fasta'):
        seq_id= seq_rec.description
        if seq_id:
            record= {'id': seq_id, 'seq': str(seq_rec.seq.lower())}
            df2=df2.append(record, ignore_index=True)
print(len(df2), df2)
```

```python
df3=pd.concat([df1, df, df2]).reset_index(drop=True)
df3
```

```python



# make sure that none of the sequences have gaps within the region of the spike
result=[]
count=0

for i, row in df3.iterrows():

    spike_result= row['seq'][spike_start:spike_end]
    
    result.append(spike_result)
        
df3['spike_seq']=result



df3
```

```python
df3['id']=df3['id'].str.replace(' ', '')
df3
```

```python
def dataframetofasta(dataframe, fastaname):
        #create fasta file
        file= fastaname + ".fasta"
        f= open(file, 'a')
        #write to the file
        for i, row in dataframe.iterrows():
            lines='>'+ row['id']+'\n'+ row['spike_seq'] + "\n"
            f.write(lines)
        f.close()


#change name of the file with every run
dataframetofasta(df3, 'genomic_analyses//SEQ//MSA//'+str(hcov)+'spike_correctMSA')

```
